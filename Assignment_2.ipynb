# Assignment 2: Reproducability and living documents

The interweb is awash with data, yet a lack of scientific computing guideliness has left most of it difficult to access. Like, really painful to access. What this means is that those able to harvest such data have a desirable skill set that, while slow at the front end and buggy to maintain (people change their websites without notice), gives one a leg up in applying for things in science after graduation. For this assignment you need to find a webpage with content you're really interested in downloading, and write a script to extract the data from it on the fly.

The purpose of this assignment is to get you to find your own data to focus on for the rest of the course. The goals here are to:

1) Come up with a research question

2) Gather and analyze appropriate data to address the data

3) Interpret your results in terms of how they address the question

The report itself should be written and analyzed within a single jupyter notebook so that it can be run from start to finish by a new user. The idea of the assignment is for you to scrape a useful dataset and do a small analysis that takes you from raw data to a figure or two.

# Layout of the report
---
Your project report should be broken into 3 or 4 sections


## Title
- An informative title that gives the reader a sense of what they are about to see

## Introduction: ~500 wds (20%)
The introduction should motivate the reader as to why they question is worth answering. In general introductions start broadly, narrow down to the focus of the paper and finish with the question(s) being addressed. 

Your grade will depend on:
- Do you provide sufficient background?
- Is the question motivated well?
- Is the question clearly stated?
- Quality of writing


## Methods/Results (60%)
Unlike a standard scientific paper I want you to have your methods and results embedded in the notebook. You can use a mixture of code and text (either code comments or markdown). Given the interactive nature of Jupyter Notebooks you can interweave your methods and results or provide one and then the other. **I do not expect a written methods and results section** separate from this section.

 - Document where and how you acquired the data
 - Document what the data are as they relate to your question
 - Include code and explanation for all analysis you conduct
     - When you include code you should explain the purpose of each analysis
     - You need to provide enough detail that you would reasonably expect another person to understand what you are doing
 - If you create tables or figures be sure to provide clear axis labels, legends and caption explaining what the figure is demonstrating. 
 
You will be graded on:
 - Clarity of documentation
 - Clarity and effectiveness of code
 - The quality of analysis - does the analysis address the question
 - Appropriateness of the data for the question
 - Quality of figures/table - do they demonstrate the main result clearly?
 - Quality of writing
 
## Discussion: ~500wds (20%)
The discussion should be an interpretation of your results. Put another way, the discussion takes the objectve results and explains what they mean in context.

- Do you address your question?
- Is your interpretation scientifically sound?
- Do you put your work in a broader context (have similar results been found in other studies?)
- Quality of writing

There are many resources for scraping data. Here are three examples
1. Ratebeer.com -
[Rate Beer](https://www.ratebeer.com) is a website devoted to beer ratings - harvest ratings for your favourites
2. fbref.com -
The [Football Reference site](https://fbref.com) is an (inferior) alternative to baseball statistics
