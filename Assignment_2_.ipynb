# Finding data
---

There are many resources for downloading data. Here are three important ones
1. Sequence Read Archive (NCBI)     
The [NCBI website](https://www.ncbi.nlm.nih.gov/sra) is a massive store of next generation sequencee data. You can search particular kinds of data (whole genome, mRNA, ChIPseq) from specific organisms. In general you can download fastq reads from there onto your own computer or the lab server
2. NCBI Genomes
The [NCBI Genome site](https://www.ncbi.nlm.nih.gov/genome/) has most of the sequenced genome and their gene annotations available for download.
3. UCSC Genome Browser
The [UCSC Genome Browswer](https://genome.ucsc.edu/) is an online tool for viewing genomes and their associated genomic annotations (tracks). In addition to viewing the genomes of many many organisms they also host a download site that allows you to download the genomes, transcriptomes, gene annotations etc etc of many organisms. 

In reality there are many many websites that hose genomic data. A little bit of searching will often get you what you want. If you are having a hard time tracking down appropriate data for your project talk to Prof. Ness or the Teaching assistant during lab

## Types of starting data
Although I will not officially constrain you to particualr types of data I want the data to be genome scale. Here are the most common forms of data that you might start with. Keep in mind you might start with only one type of data or you might start with multiple. It will depend on your question.

1. Fastq reads from a next generation sequencing (whole genome, RNAseq, CHiPseq, ddRAD etc, microbiome)
2. Fasta files of complete genomes, transcriptomes, proteins
3. VCF files - annotations of variation in a particualr genome 
4. BAM files - genome alignments usually of NGS data.
